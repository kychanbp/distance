```r
fiat_hr_poly <-ds(fiat, key = 'hr', adjustment = 'poly', truncation = '5%')
```

```r
#print out the summary statistics of hazard rate with cosine adjustment
summary(fiat_hr_poly)
```

```r
#plot
par(mfrow=c(1,2)) #divide the figure windown to 1*2
plot(fiat_hr_poly, main="Duck nests, harzard rate poly detection function") #plot detection function
fit.test <- ddf.gof(fiat_hr_poly$ddf) #plot the QQ plot to fitness function
```
